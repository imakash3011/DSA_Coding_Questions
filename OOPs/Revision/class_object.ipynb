```python
# 1. Write a python program to Create Student Class

class Student:
    id = 0
    name = ""
    gender = ""
    total = 0
    per = 0

    def setData(self, id, name, gender, total, per):
        self.id = id
        self.name = name
        self.gender = gender
        self.total = total
        self.per = per
    
    def showData(self):
        print("Student id is:", self.id)
        print("name of the student : ", self.name)
        print("gender of the student :", self.gender)
        print("total score of the student : ", self.total)
        print("Percentage of the student :", self.per)

s1 = Student()
s2 = s1.setData(1, "akash", "male", "500", "100")
s1.showData()
```

```python
# Write a python program to Create Student Class with Constructor and Destructor

class Student:
    def __init__(self):
        # Use self. to define instance attributes
        self.id = 0
        self.name = ""
        self.gender = ""
        self.total = 0
        self.per = 0

    def __del__(self):
        print(f"Object for {self.name} Destroyed")

    def setData(self, id, name, gender, total, per):
        self.id = id
        self.name = name
        self.gender = gender
        self.total = total
        self.per = per
    
    def showData(self):
        print(f"ID: {self.id} | Name: {self.name} | Total: {self.total}")

# 1. Create the object
s1 = Student()
s1.setData(1, "Akash", "Male", "500", "100")
s1.showData()

# 2. How to "call" the destructor
print("--- Attempting to delete s1 ---")
del s1 

# The code below would now fail because s1 is gone
# s1.showData()
```

```python
# 3. Write a python program to implement Getters and Setters in a class

class Student:
    def __init__(self):
        self.name = ""
        self.total = 0
        self.per = 0
    
    def setName(self, name):
        self.name = name
    def getName(self):
        return self.name
    
    def setTotal(self, total):
        self.total = total
    def getTotal(self):
        return self.total
    
    def setPercent(self, per):
        self.per = per
    def getPercent(self):
        return self.per
    
stu1 = Student()
stu1.setName("Akash")
print(stu1.getName())

stu1.setTotal(999)
print(stu1.getTotal())

stu1.setPercent(99.9)
print(stu1.getPercent())
```

```python
# 4. Write a python program to Implement Abstraction using Abstract class

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def accelerate(self, speed):
        pass

class Car(Vehicle):

    def accelerate(self, speed):
        print("Car is accelerating at:", speed)

c1 = Car()
c1.accelerate(99)
```

```python
# Write a python program to check Palindrome number using object oriented approach

class Palindromic:
    def __init__(self,str1):
        self.str1 = str1
    
    def check_palindromic(self, str1):
        rev_str1 = str1[::-1]
        return rev_str1==str1
    
p1 = Palindromic("mom")
print(p1.check_palindromic("mom"))
```

```python
## Bank Management System

class Bank:

    def __init__(self):
        self.account= {}
    
    def create_account(self, account_number, name, balance):
        if account_number not in self.account:
            self.account[account_number]= {'name':name, 'balance':balance}
        else:
            return "Account Already Exists!"
    
    def deposit(self, account_number, amount):
        if account_number in self.account:
            self.account[account_number]['balance'] += amount

    def show_details(self):
        for  i, j in self.account.items():
            print(i, j)


a1 = Bank()
a1.create_account(123, "Akash", 100)
# a1.create_account(456, "Divanshu", 200)
a1.show_details()

a1.deposit(123, 5000)
a1.show_details()
```
