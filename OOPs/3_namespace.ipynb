```python
## Decorators 

# A decorator in python is a function that receive another function 
# as input and adds some functionality (decoration) to and it and returns it.

# This can happen only because python functions are 1st class citizens
# There are 2 types of decorators available in python
# 1. Built in decorators like @staticmethod, @classmethod, @abstract and @property etc
# 2. user defined decorator
```

```python
# python functions are 1st class function (you can do any sort of operation means function here)

def func():
    print("hello")

a = func()
a
```

```python
def modify(func, num): ## taking function and a number
    return func(num) ## return function by passing that value in it

def square(num):
    return num**2

modify(square, 2)
```

```python
#Simple example
## Even after parent class usage is not there but child class will be able to use it (decorator is based on this concept)
def my_decorator(func):
    def wrapper():
        print("*****************")
        func()
        print("*********************")
    return wrapper

def hello():
    print("hello")


a = my_decorator(hello)
a()
```

```python
# more functions


def my_decorator(func):
    def wrapper():
        print("*****************")
        func()
        print("*********************")
    return wrapper

@my_decorator  ## just a shortcut to use decorator so that you won't need to call is indirectly
def hello():
    print("hello")


hello()
```

```python
## Anything meaningful
# decorator that will help you to print the execution time of function

import time

## This is the decorator func
def timer(func): 
    def wrapper(): ## wrapper(*args) if you want to pass argument
        start = time.time()
        func() # func(*args)
        print('time taken by', func.__name__, time.time() - start, 'secs')
    return wrapper

@timer
def hello():
    print("hello world")
    time.sleep(2)

@timer
def sqaure(num): ## this code will not work until you add argument in decorator function
    time.sleep(1)
    return num**2


hello()
```

```python
def sanity_check(data_type):
    def outer_wrapper(func):
        def inner_wrapper(*args):
            if type(args[0]) == data_type:
                func(*args)
            else:
                raise TypeError("ye datatype ni chlega")
        return inner_wrapper
    return outer_wrapper

@sanity_check(int)
def squre(num):
    print(num**2)


@sanity_check(str)
def greet(name):
    print("hello", name)

square(2)
# greet("akash")
```
