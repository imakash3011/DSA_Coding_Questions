```python
# Largest Element

def largest_num(nums):
    val = float('-inf')
    for i in nums:
        if i>val:
            val = i
    return val


nums =  [3, 3, 0, 99, -40]
print(largest_num(nums))
```

```python
## Union of two sorted array

class Solution:
    def unionArray(self, nums1, nums2):
        left = 0
        right = 0
        m = len(nums1)
        n = len(nums2)
        result = []
        while left<m and right<n:
            if nums1[left] == nums2[right]:
                if not result or result[-1] != nums1[left]:
                    result.append(nums1[left])
                left+=1
                right+=1
            elif nums1[left] < nums2[right]:
                if not result or result[-1] != nums1[left]:
                    result.append(nums1[left])
                left+=1
            else:
                if not result or result[-1] != nums2[right]:
                    result.append(nums2[right])
                right+=1
        
        while left<m:
            if not result or result[-1] != nums1[left]:
                result.append(nums1[left])
            left+=1
        
        while right<n:
            if not result or result[-1] != nums2[right]:
                result.append(nums2[right])
            right+=1

        return result

nums1 = [3, 4, 6, 7, 9, 9]
nums2 = [1, 5, 7, 8, 8]
c1 = Solution()
print(c1.unionArray(nums1, nums2))
```

```python
# Longest subarray with sum K (use prefix sum)

## OPTIMAL



##BRUTEFORCE
class Solution:
    def longestSubarray(self, nums, k):
        
        final_res = 0
        n = len(nums)
        for i in range(n):
            cnt = 1
            temp = nums[i]
            for j in range(i+1,n):
                temp += nums[j]
                cnt+=1
                if temp==k:
                    final_res = max(final_res, cnt)              
        return final_res


nums = [10, 5, 2, 7, 1, 9]  
k=15
c1 = Solution()
print(c1.longestSubarray(nums, k))
```

```python
# Two Sum

class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        result = {}
        for i in range(len(nums)):
            remaining = target - nums[i]
            if remaining in result:
                return [result[remaining], i]
            
            result[nums[i]] = i
        return [-1, -1]

nums = [3,3]
target = 6
c1 = Solution()
print(c1.twoSum(nums, target))
```

```python
# Maximum Subarray

class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        max_val = 0
        temp =0
        for i in nums:
            temp+= i
            if temp<0:
                temp =0
            else:
                max_val = max(max_val, temp)

        return max_val

nums =[5,4,-1,7,8]
c1 = Solution()
print(c1.maxSubArray(nums))
```

```python
# Longest Consecutive Sequence in an Array


class Solution:
    def longestConsecutive(self, nums):
        nums.sort()
        max_val = 1
        cnt=1
        for i in range(len(nums)-1):
            if nums[i]+1 == nums[i+1]:
                cnt+=1
                max_val = max(max_val, cnt)
            elif nums[i] == nums[i+1]:
                continue
            else:
                cnt = 1
        return max_val


nums = [100, 4, 200, 1, 3, 2]
c1 = Solution()
print(c1.longestConsecutive(nums))
```

```python
# Rotate matrix by 90 degrees

class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        row = len(matrix)
        col = len(matrix[0])

        for i in range(row):
            for j in range(i+1, col):
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
        
        for  i in range(row):
            matrix[i].reverse()
        
        return matrix


matrix = [[1,2,3],[4,5,6],[7,8,9]]
c1 = Solution()
print(c1.rotate(matrix))
```

```python
# Pascal's Triangle

class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
        final_result = []
        for i in range(numRows):
            row = [1]*(i+1)

            if i>=2:
                prev = final_result[i-1]
                for j in range(1, i):
                    row[j] = prev[j-1] + prev[j]

            final_result.append(row)
        return final_result

numRows = 5
c1 = Solution()
print(c1.generate(numRows))
```
