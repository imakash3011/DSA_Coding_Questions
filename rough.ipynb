```python
# x = 123

class Solution:
    def reverse(self, x: int) -> int:
        result = 0
        while x>0:
            rem = x%10
            result = result*10+rem
            x = x//10
        return result
    

s1 = Solution()
print(s1.reverse(123))
```

```python
# lst = [[1], [1,1], [1,1,1]]


nums =[1,1]
m = len(nums)-1
left, right = 0,1
new_lst = [nums[left]]
while left<=m:
    new_lst.append(nums[left]+nums[right])
    if right == m:
        left+=1
        new_lst.append(nums[right])
        break
    else:
        left+=1
        right+=1   
print("final", new_lst)
```

```python
lst = [[1], [1,1], [1,1,1]]
final_output = []
for nums in lst:
    m = len(nums)-1
    if m==0:
            final_output.append(nums)
            continue
    left, right = 0,1
    new_lst = [nums[left]]
    while left<=m:
        new_lst.append(nums[left]+nums[right])
        if right == m:
            left+=1
            new_lst.append(nums[right])
            break
        else:
            left+=1
            right+=1   
    final_output.append(new_lst)
print("final", final_output)
```
