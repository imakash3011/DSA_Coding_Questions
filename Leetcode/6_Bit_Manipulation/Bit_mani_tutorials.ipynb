```python
# Decimal to Binary

def convert2binary(num):
    res =[]

    while num>0:
        reminder = num%2
        res.append(reminder)
        num = num//2
    return res[::-1]


print(convert2binary(13))
```

```python
# Binary to Decimal 

def binary_to_decimal(num):
    res = 0
    new_num = str(num)[::-1]

    for i in range(len(new_num)):
        res += int(new_num[i]) * (2 ** i)

    return res

print(binary_to_decimal(1101))
```

```python
def demo_bitwise_operators(a: int, b: int):
    # Format the numbers as 8-bit binary strings for visual clarity
    binary_a = f"{a:08b}"
    binary_b = f"{b:08b}"
    
    print(f"Inputs:")
    print(f"a = {a} (Binary: {binary_a})")
    print(f"b = {b} (Binary: {binary_b})")
    print("-" * 40)
    
    # 1. Bitwise AND (&)
    # Rules: 1 & 1 = 1 | 1 & 0 = 0 | 0 & 0 = 0 (Only 1 if BOTH are 1)
    and_result = a & b
    print("Bitwise AND (a & b):")
    print(f"  {binary_a}")
    print(f"& {binary_b}")
    print(f"  {'-' * 8}")
    print(f"  {and_result:08b}  -> Decimal: {and_result}")
    print("\n" + "-" * 40 + "\n")
    
    # 2. Bitwise OR (|)
    # Rules: 1 | 1 = 1 | 1 | 0 = 1 | 0 | 0 = 0 (1 if AT LEAST ONE is 1)
    or_result = a | b
    print("Bitwise OR (a | b):")
    print(f"  {binary_a}")
    print(f"| {binary_b}")
    print(f"  {'-' * 8}")
    print(f"  {or_result:08b}  -> Decimal: {or_result}")

# Run the demo with two numbers
# 12 in binary is 00001100
# 25 in binary is 00011001
demo_bitwise_operators(12, 25)
```

```python
## Bit set or not

# USING LEFT SHIFT
def check_set_bit(num, i):
    mask = 1<<i
    if (num & mask)!=0:
        return True
    return False

print(check_set_bit(0b1101, 3))
print(check_set_bit(0b1101, 1))

```

```python
## Bit set or not

## USING RIGHT SHIFT

def check_set_bit_right_shift(num, i):
    # Shift the number to the right by 'i' positions.
    # This brings the i-th bit all the way to the rightmost spot.
    shifted_num = num >> i
    
    # Check if the rightmost bit is 1 by ANDing it with 1
    if (shifted_num & 1) == 1:
        return True
    return False

# 0b1101 has: Bit 3 = 1, Bit 2 = 1, Bit 1 = 0, Bit 0 = 1
print(check_set_bit_right_shift(0b1101, 3))  # Output: True
print(check_set_bit_right_shift(0b1101, 1))  # Output: False
```

```python
## SET THE ith Bit  (using left shift)

def set_ith_bit(num, i):
    mask = 1<<i
    return f'{num|mask:b}'   ## Return the result as a binary string


num = 0b1101
i = 1
print(set_ith_bit(num,i ))
```

```python
## CLEAR the ith bit


## SET THE ith Bit (means make it to 0)

def clear_ith_bit(num, i):
    mask = ~(1<<i)

    # print(bin(mask))
    # print(bin(~mask))   ## Note: Python prints negative binary numbers with a minus sign due to its infinite precision system, but underneath, it represents the inverted bit structure
    
    return f'{num&mask:b}' 

num = 0b1101
i = 3
print(clear_ith_bit(num,i ))
```

```python
## Toggle ith bit

def toggle_ith_bit(num, i):
    mask = 1<<i
    return f'{num^mask:b}'   ## Return the result as a binary string

num = 0b1101
i = 2
print(toggle_ith_bit(num,i ))
```

```python
## Check if no.  is power of 2 or not

def check_pow_of_2(num):
    if num<=0:
        return False
    if num & (num - 1)== 0:
        return True
    return False

num = 0
print(check_pow_of_2(num))
```

```python
## Number of bit need to be flipped to convert A to B

def convert_A_B(a,b):
    xor_a_b = a^b  ## XOR will make those indexes 1 where flip is required because you know same will show 0 and diff bit show 1 in xor
    cnt = 0 
    for i in range(0, 32):
        if xor_a_b & (1<<i)!=0:  ## Now taking mask using (1<<i) and checking all the bits
            cnt+=1
    return cnt

a = 10
b = 7
print(convert_A_B(a, b))
```

```python
## Find single number

def find_single_num(arr):
    res = 0
    for i in arr:  # XOR every number in the array
        res ^=i
    
    return res

arr =[5, 1,3,3,7,1,7]
print(find_single_num(arr))
```

```python
## print all the subset

def print_all_subset(nums):
    n = len(nums)
    total_subset = 1<<n               # or 2**n 
    all_subsets = []

    for i in range(total_subset):     # Loop from 0 to 2^n - 1
        current_subset = []
        
        for j in range(n):            # Check each of the n bit positions
            mask = 1 << j             # Create a mask with a 1 at the j-th position
            if (i & mask) != 0:       # If the bitwise AND is not 0, it means the j-th bit is set
                current_subset.append(nums[j])
        all_subsets.append(current_subset)
    return all_subsets

nums = [1,2,3]
print(print_all_subset(nums))
```

```python
## Addition 

x = 7
add_one = - (~x)
print(f"7 + 1 = {add_one}")  # Output: 8
# Bitwise breakdown:
# 7  in binary: 0000 0111
# ~7 in binary: 1111 1000 (which is -8 in decimal)
# -(-8) = 8
```

```python
## Substraction

x = 7
# Loop-based pure bitwise subtraction
borrow = 1
sub_one = x
while borrow != 0:
    borrows = (~sub_one) & borrow
    sub_one = sub_one ^ borrow
    borrow = borrows << 1

print(f"7 - 1 = {sub_one}")  # Output: 6
```

```python
## Multiplication

x = 7
a = x << 1  # 7 * 2 = 14
b = x       # 7

# Pure bitwise addition (Half Adder loop)
while b != 0:
    carry = a & b
    a = a ^ b
    b = carry << 1

mul_three = a
print(f"7 * 3 = {mul_three}")  # Output: 21
```

```python
## Division

dividend = 7
divisor = 3

quotient = 0
# We align the divisor to the highest possible bit position
# For small numbers like 7, shifting by 4-5 bits is plenty to check
for i in range(31, -1, -1):
    # Check if (divisor << i) <= dividend purely using bitwise logic
    # To keep it readable, we check if the shifted divisor fits
    if (divisor << i) <= dividend:
        # quotient |= (1 << i)
        quotient = quotient | (1 << i)
        
        # dividend -= (divisor << i) using bitwise subtraction loop
        sub_val = divisor << i
        while sub_val != 0:
            borrows = (~dividend) & sub_val
            dividend = dividend ^ sub_val
            sub_val = borrows << 1

print(f"7 // 3 = {quotient}")  # Output: 2
```
