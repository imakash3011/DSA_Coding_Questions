```python
def count_down(n):
    # Initialize the "static" variable if it doesn't exist yet
    if not hasattr(count_down, "counter"):
        count_down.counter = 0  
    
    # Increment the static variable every time the function is called 
    count_down.counter += 1 
    print(count_down.counter) 
    
    # Base case to stop recursion
    if n <= 0:
        print("Blast off!")
        return
    
    # Recursive step
    count_down(n - 1)

# --- Running the code ---
count_down(3)

# Check the static variable after the function finishes
print(f"The function was called {count_down.counter} times.")
```

```python
def sum_parameterized(n, current_sum):
    # Base Case: When n reaches 0, we have our final answer in current_sum
    if n == 0:
        print(f"Final Sum: {current_sum}")
        return
    
    # Recursive Call: Pass the updated sum down to the next call
    sum_parameterized(n - 1, current_sum + n)

# Usage
sum_parameterized(3, 0)
```

```python
def sum_functional(n):
    # Base Case: The sum of 0 is 0
    if n == 0:
        return 0
    
    # Recursive Call: Wait for the result of (n-1) and then add n to it
    return n + sum_functional(n - 1)

# Usage
result = sum_functional(3)
print(f"Final Sum: {result}")
```

```python
## Fibonacci using multiple recursion call

# def fibo(n, left, right):
#     if right>n:
#         return right
#     print(right)
#     return fibo(n, right, left+right)


def fibo(n):
    # Base cases: F(0) = 0, F(1) = 1
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    # Multiple recursion: Two recursive calls in a single return statement
    return fibo(n - 1) + fibo(n - 2)

# Example usage: Get the 6th Fibonacci number
# Define how many elements of the sequence you want to print
n_terms = 7 

# Loop through and print each number in the sequence
for i in range(n_terms):
    print(fibo(i), end=" ")
```

```python
# Generate Subsequences with Sum K

class Solution:
    def subsets_sum_k(self, nums: List[int], target:int) -> List[List[int]]:
        result = []

        def helper_func(idx,subset, target):
            if idx >= len(nums):
                # print(subset)
                if sum(subset) == target:
                    result.append(subset.copy())
                return

            subset.append(nums[idx])
            helper_func(idx+1,subset, target)

            subset.pop()
            helper_func(idx+1,subset, target)
        
        helper_func(0, [] , target)
        return result
            

nums = [3,5,6,7,2,1] 
target = 9
c1 = Solution()
print(c1.subsets_sum_k(nums, target))
```
